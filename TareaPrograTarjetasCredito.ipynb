```python
import pandas as pd
import matplotlib.pyplot as plt

archivo = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",")
```

```python
procesoG = pd.crosstab(index=archivo["Gender"],columns="frecuencia")
print(procesoG)
gender = procesoG.loc[procesoG.index]["frecuencia"]
plt.pie(gender, labels=gender.index)
plt.xlabel("Genero por tarjetas")

plt.show

procesoG.to_excel("Tajetas_por_genero.xlsx")
```
